Интернет-магазин «Стримчик» продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

Предстоит подготовить данные, предоставленные до 2016 года, провести исследовательский анализ, составить портрет пользователя каждого региона, сформулировать и проверить гипотезы для получения требуемых результатов.

```python
from scipy import stats as st
import math as mt
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# создадим датафрейм
try:
    df = pd.read_csv('datasets/games.csv')

except:
    df = pd.read_csv('/datasets/games.csv')
# изучим общую информацию
df.info()
```

Итак, изучили общую информацию по имеющимся данным. Данные содержат информацию об играх, выпущенных до 2016 года, в том числе данные по продажам игры, по оценкам от пользователей и критиков и т.д.

Предварительно можно сказать, что данных вполне достаточно для проведения анализа и решения поставленной задачи. Но в датафрейме присутствуют проблемы, которые надо устранить, то есть предобработать данные.

## Предобработка данных

### Изменение типов данных и обработка пропусков

Начнем подготовку данных с того, что изменим названия столбцов в соответствии с хорошим стилем.

В названиях столбцов лучше использовать змеиный регистр, написание без заглавных букв и тд, в наших же данных полно заглавных букв, необходимо это исправить.

```python
df.columns
```

```python
df.columns = df.columns.str.lower()
```

Далее преобразуем данные в нужные типы.

```python
df.info()
```

Как видно из информации по датафрейму, тип данных неправильно установлен в большинстве столбцов, кроме столбцов с продажами, где уже указан тип float. 

Начнем по порядку со столбцов, где необходим строчный тип (название, жанр, платформа, возрастная категория).

```python
# установим строчный тип
df['name'] = df['name'].astype('string')
df['platform'] = df['platform'].astype('string')
df['genre'] = df['genre'].astype('string')
df['rating'] = df['rating'].astype('string')
```

Обработаем пропуски в столбце с годом выпуска на -1 (год выпуска не указан) и изменим тип данных на int.

```python
df['year_of_release'] = df['year_of_release'].fillna(-1)
df['year_of_release'] = df['year_of_release'].astype('int')
```

Заполним пропуски столбцах с жанром, именем и возрастной категорией.

```python
df['rating'] = df['rating'].fillna('unknown')
df['genre'] = df['genre'].fillna('unknown')
df['name'] = df['name'].fillna('unknown')
```

В столбце с пользовательской оценкой заполним пропуски, мешающие изменить тип данных, на -1, что будет означать, что оценки нет или что указано 'tbd'. После этого изменим тип данных на float.

```python
df['user_score'] = df['user_score'].fillna(-1)
df['user_score'] = df['user_score'].replace('tbd', -1)
df['user_score'] = df['user_score'].astype('float')
```

Также поступим и с этим столбцом, сначала заполним пропуски на -1, потом поменяем тип данных на int.

```python
df['critic_score'] = df['critic_score'].fillna(-1)
df['critic_score'] = df['critic_score'].astype('int')
```

```python
df.info()
```

### Вывод

Итак, типы данных столбцов датафрейма изменены на нужные. В столбцах **name, platform, genre, rating** это строчный тип *string*. В столбце **critic_score** это целочисленный *int*, так как все оценки от критиков - это целые числа, а в столбце **user_score** - тип *float*, так как оценки пользователей - числа с плавающей запятой.

Также заполнены некоторые пропуски в столбцах с рейтингом. Использовать метод с медианным или средним значением было бы некорректно, так как оценка - субъективная величина. Поэтому пропуски были заменены на значение -1, чтобы была возможность изменить тип данных без ошибок. Такой способ не повлияет на дальнейшие результаты, но упростит работу с данными. Также в столбце **user_score** значение **"tbd"** заменено на -2 по тому же принципу.

Пропуски в данных могли появиться из-за технических проблем (ошибки при переносе, редактировании, копировании данных), человеческого фактора (кто-то случайно или умышленно не ввел все данные) или если датасет собран из нескольких источников с разными наборами параметров, что после соединения привело к потере некоторых данных.

### Расчёты и добавление результатов в таблицу

Посчитаем суммарные продажи во всех регионах и запишем их в отдельный столбец.

```python
df['sum_sales'] = df['na_sales'] + df['jp_sales'] + df['eu_sales'] + df['other_sales']
df.head()
```

### Вывод

В таблицу был добавлен столбец с суммарными продажами игры в каждом регионе. Теперь можно будет использовать его в дальнейшем анализе.

## Исследовательский анализ данных

Имея уже предобработанные данные можно провести их анализ.

Сначала посмотрим, сколько игр выпускалось в разные годы.

```python
df.groupby('year_of_release')['name'].count()
```

Резкий рост количества видеоигр начался во второй половине 90-х. До 1994г количество выпущенных игр не превышало ста. В целом, можно предположить, что данные до примерно 2000-х не существенны для исследования.

### Топ-10 платформ 

Посмотрим, как менялись продажи по платформам. Выберем платформы с наибольшими суммарными продажами и построим распределение по годам. 

```python
# топ-10 платформ с наибольшими продажами 
pf = df.groupby('platform')['sum_sales'].sum().sort_values(ascending=False).head(10)
pf
```

```python
# датафрейм с данными по играм, выпущенными на платформах из топ-10
df_platforms_top = df.query('platform in @pf.index')
df_platforms_top.head()
```

Получился датафрейм только из игр на платформах из топ-10. Теперь построим распределение по годам.

```python
# построим столбчатые диаграммы распределения продаж по годам и платформам (-1 на оси x означает, что год не указан)
for platform in pf.index:
    df_platforms_top[df_platforms_top['platform'] == platform].pivot_table(index='year_of_release', 
                                                                   values='sum_sales', 
                                                                   aggfunc='mean').plot(kind='bar',
                                                                                        title='Распределение по годам продаж игр на платформе ' + platform, 
                                                                                        xlabel='Год выпуска игр на платформе ' + platform, 
                                                                                        ylabel='Сумма продаж за год (млн проданных копий)')
    plt.show()
```

Итак, исходя из графиков: 

- PS2 имеет данные об играх, выпущенных с 2000 по 2011 (11 лет)
- X360 имеет данные об играх, выпущенных с 2005 по 2016г, хотя игры и дальше могут появляться (так как данных после 2016г нет), но видно, что продажи резко падают, так что можно сказать, что платформа уже устарела (11 лет)
- PS3 имеет продажи с 2006 по 2016г, также видно резкое падение продаж в 2016г (10 лет)
- Wii - продажи с 2006 по 2016 год, была крайне популярна в год выхода(2006), но падение продаж идет сразу с 2007г, в 2016 продажи уже минимальны (10 лет)
- DS - продажи с 2004 по 2013г, платформа появилась в 2004г, 1985г возможно появился по ошибке (в итоге 9 лет)
- PS - продажи с 1994 по 2003г (9 лет)
- GBA - продажи с 2000 по 2007г (7 лет)
- PS4 - продажи с 2013 по 2016г, но платформа достаточно новая, так что чтобы говорить о конце срока "жизни" нельзя
- PSP - продажи с 2004 по 2015г (11 лет)
- PC - это собственно компьютер, так что нельзя сказать о конкретном сроке "жизни"

В итоге, можно сказать, что платформы в среднем "живут" около 10 лет. Можно сделать вывод, что часто бывает, что основные продажи на определенной платформе происходят в первые года 3-4, потом платформа еще "жива", но гораздо менее популярна и игр становится все меньше.

Также можно увидеть, что данные до 2000г почти не попали в это исследование. Но нам нужен *актуальный период* для прогнозирования на следующий 2017 год. Для такой динамично меняющейся индустрии он должен быть не более нескольких лет. Возьмем как *актуальный период* данные с 2014г.

Также построим общий для платформ график.

```python
df = df[df['year_of_release'] >= 2005]
```

```python
# актуальный период
df_actual = df[df['year_of_release'] >= 2014]
df_actual.pivot_table(index=['year_of_release', 'platform'], 
                         values='sum_sales',
                         aggfunc='mean').unstack()
```

В итоге, к 2016г можно выделить как потенциально прибыльные только относительно новые PS4, XOne, WiiU, а также 3DS. Но еще, если логически подумать, можно выделить и PC, так как игры на компьютере будут выпускать и дальше, следовательно и продажи никуда не исчезнут.

```python
df_actual.pivot_table(index=['year_of_release', 'platform'], 
                         values='sum_sales', 
                         aggfunc='mean').unstack().boxplot(figsize=(16,12), grid=True)
plt.title('Распределение глобальных продаж игр в разбивке по платформам')
plt.xlabel('Платформа')
plt.ylabel('Сумма продаж за год (млн проданных копий)')
plt.show()
```

### Связь между отзывами и продажами игр

## Проверка гипотез

Далее проверим гипотезы:
- Средние пользовательские рейтинги платформ Xbox One и PC одинаковые
- Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные

### Гипотеза о равенстве средних пользовательских рейтингов платформ Xbox One и PC

Нулевая гипотеза всегда должна означать равенство. Таким образом, нулевая гипотеза будет гласить: средние пользовательские рейтинги платформ Xbox One и PC равны.

Альтернативная гипотеза: средние пользовательские рейтинги платформ Xbox One и PC различны.

```python
# критический уровень статистической значимости
alpha = .05
# проверка гипотезы о равенстве средних пользовательских оценок игр на платформах XBox One и PC 
# у нас нет оснований полагать, что дисперсии выборок различны, так что параметр equal_var опустим
results = st.ttest_ind(df['user_score'][df['platform'] == 'XOne'], df['user_score'][df['platform'] == 'PC']) 
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Итак, проверка показала, что нулевая гипотеза отвергнута. Следовательно, вероятность того, что средние пользовательские рейтинги платформ Xbox One и PC отличаются, выше. Значит, пользовательские рейтинги игр действительно зависят от платформ.

### Гипотеза о различии средних пользовательских рейтингов жанров Action и Sports.

Нулевая гипотеза: средние пользовательские рейтинги жанров Action и Sports равны.

Альтернативная гипотеза: средние пользовательские рейтинги жанров Action и Sports различны.

```python
# критический уровень статистической значимости
alpha = .05
# проверка гипотезы о равенстве средних значений пользовательских оценок игр жанра экшн и спорт
# здесь также нет оснований полагать, что дисперсии выборок различны, так что параметр equal_var опустим
results = st.ttest_ind(df['user_score'][df['genre'] == 'Action'], df['user_score'][df['genre'] == 'Sports']) 
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

В результате нулевая гипотеза отвергнута. Следовательно, вероятность того, что средние пользовательские рейтинги игр в жанре экшн и спорт отличаются, выше. Значит, пользовательские рейтинги игр зависят от жанра.
